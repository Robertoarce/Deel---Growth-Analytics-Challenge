# Data Extraction

```python
#load
paid = pd.read_csv("./../data/paid_platform.csv")
conversion = pd.read_csv("./../data/conversion_data.csv")

#format
paid.rename(columns={col: col.lower().replace('# ', '').replace(' ', '_').replace('$_', '').replace('utm_', '') for col in paid.columns}, inplace=True)
conversion.rename(columns={col: col.lower().replace('# ', '').replace(' ', '_').replace('$_', '').replace('utm_', '') for col in conversion.columns}, inplace=True)


#Change type
paid['cost'] = paid.cost.str.replace("$",'').str.replace(',','').astype(float)
```

# Data Export

No utm-source nor medium should have numbers in the final usage table. (Numbers as keys for other table's values should be done beforehand.)

```python
m_medium = conversion.medium.isin(conversion_mediums_filtered)
m_source = conversion.source.isin(conversion_source_filtered)
conversion = conversion[m_medium & m_source]
```

# JOINTURE

```python
paid_filtered = paid.groupby(['medium', 'source', 'traffic_week'],as_index= False).sum(['impressions', 'clicks', 'cost'])
paid_filtered.head(3)
```

```python
conversion_filtered = conversion.groupby(['touched_date',  'source'], as_index=False).sum(['demo_occurred',
       'sales_qualified_opportunity', 'closed_won_opportunity'])
conversion_filtered.head(3)
```

```python
df = pd.merge(paid_filtered, conversion_filtered,left_on=['traffic_week','source'], right_on=['touched_date','source'], how='left',suffixes=('_paid', '_conversion'))
df.head(2)
```

```python
mask = df['touched_date'] != df['traffic_week']
df_filtered = df[~mask]
```
